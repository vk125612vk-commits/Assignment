# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage:
c = Circle(5)
r = Rectangle(4, 6)
print(f"Circle Area: {c.area():.2f}")
print(f"Rectangle Area: {r.area()}")
```

# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, make):
        super().__init__(vehicle_type)
        self.make = make

class ElectricCar(Car):
    def __init__(self, vehicle_type, make, battery_size):
        super().__init__(vehicle_type, make)
        self.battery = battery_size

# Example usage:
tesla = ElectricCar("Sedan", "Tesla", "100kWh")
print(f"Type: {tesla.type}, Make: {tesla.make}, Battery: {tesla.battery}")
```

# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

```python
# Base class
class Bird:
    def fly(self):
        print("Some birds can fly, some cannot.")

# Derived class 1
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky.")

# Derived class 2
class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly, but swims in water.")

# Polymorphism in action
def make_bird_fly(bird):
    bird.fly()

# Creating objects
sparrow = Sparrow()
penguin = Penguin()

# Calling the same method on different objects
make_bird_fly(sparrow)
make_bird_fly(penguin)
```

# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient funds")

    def check_balance(self):
        return self.__balance

# Example usage:
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(f"Current Balance: {account.check_balance()}")
```

# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

```python
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Example usage:
print(f"Addition: {MathOperations.add_numbers(10, 5)}")
print(f"Subtraction: {MathOperations.subtract_numbers(10, 5)}")
```

# 8. Implement a class Person with a class method to count the total number of persons created.

```python
class Person:
    """
    Represents a person and counts the total number of instances created.
    """
    # A class variable to store the total count of Person instances
    total_persons = 0

    def __init__(self, name):
        """
        Initializes a new Person instance and increments the total count.
        """
        self.name = name
        Person.total_persons += 1

    @classmethod
    def get_total_persons(cls):
        """
        A class method that returns the total number of Person instances created.

        Args:
            cls: The class itself (Person).

        Returns:
            int: The total count of Person instances.
        """
        return cls.total_persons

# Example usage:
p1 = Person("Alice")
p2 = Person("Bob")
p3 = Person("Charlie")

# Using the class method to get the total count
count = Person.get_total_persons()
print(f"Total number of persons created: {count}")

# The class method can also be called from an instance (though typically called on the class)
count_from_instance = p1.get_total_persons()
print(f"Total number of persons from an instance call: {count_from_instance}")
```

# 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".

```python
class Fraction:
    def __init__(self, numerator, denominator):
        """
        Initializes a Fraction object with a numerator and denominator.
        Ensures the denominator is not zero.
        """
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        """
        Overrides the standard string representation of the object.
        Displays the fraction in "numerator/denominator" format.
        """
        return f"{self.numerator}/{self.denominator}"

# Example Usage:
# Create an instance of the Fraction class
my_fraction = Fraction(3, 4)

# Print the fraction object; the __str__ method is called automatically
print(my_fraction)

# Another example
another_fraction = Fraction(1, 2)
print(another_fraction)
```

# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Example usage:
v1 = Vector(2, 4)
v2 = Vector(3, 1)
v3 = v1 + v2
print(f"Resultant Vector: {v3}")
```

# 12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

```python
class Student:
    def __init__(self, name, grades):
        """
        Initializes a Student object with a name (string) and a list of grades (list of numbers).
        """
        self.name = name
        self.grades = grades

    def average_grade(self):
        """
        Computes and returns the average of the grades.
        Returns 0 if the grades list is empty.
        """
        if not self.grades:
            return 0
        # The sum() function adds up all the numbers in the list.
        # The len() function returns the number of items in the list.
        return sum(self.grades) / len(self.grades)

# --- Example Usage ---

# Create a student instance
student1 = Student("Vijay", [85, 90, 78, 92])

# Calculate and print the average grade
average = student1.average_grade()
print(f"{student1.name}'s grades: {student1.grades}")
print(f"{student1.name}'s average grade is: {average:.2f}")

# Example with an empty list of grades
student2 = Student("Vikki", [])
average_bob = student2.average_grade()
print(f"\n{student2.name}'s grades: {student2.grades}")
print(f"{student2.name}'s average grade is: {average_bob:.2f}")
```

# 13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.

```python
class Rectangle:
    """
    Represents a rectangle and provides methods to set its dimensions and calculate its area.
    """

    def __init__(self, width=0, height=0):
        """
        Initializes the Rectangle with optional initial width and height.
        """
        self.width = width
        self.height = height

    def set_dimensions(self, width, height):
        """
        Sets the width and height of the rectangle.

        Args:
            width (int or float): The width of the rectangle.
            height (int or float): The height of the rectangle.
        """
        if width <= 0 or height <= 0:
            raise ValueError("Dimensions must be positive values.")
        self.width = width
        self.height = height

    def area(self):
        """
        Calculates and returns the area of the rectangle.

        Returns:
            int or float: The area of the rectangle (width * height).
        """
        return self.width * self.height

# --- Example Usage ---
# Create an instance of the Rectangle class
my_rectangle = Rectangle()

# Set the dimensions using the set_dimensions method
my_rectangle.set_dimensions(5, 10)

# Calculate and print the area using the area method
rectangle_area = my_rectangle.area()
print(f"The area of the rectangle with dimensions {my_rectangle.width}x{my_rectangle.height} is: {rectangle_area}")

# You can also create an instance with initial dimensions
another_rectangle = Rectangle(3, 4)
print(f"The area of the another rectangle is: {another_rectangle.area()}")

# Setting new dimensions for the same instance
my_rectangle.set_dimensions(7, 2)
print(f"The new area of the first rectangle is: {my_rectangle.area()}")
```

# 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours workedand hourly rate. Create a de rived class Manager that adds a bonus to the salary.

```python
class Employee:
    """
    Represents a generic employee and calculates salary based on hours worked and hourly rate.
    """
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        """
        Calculates the base salary.
        """
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    """
    Represents a manager, a type of employee who receives an additional bonus.
    """
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        # Call the parent class constructor to handle common attributes
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        """
        Overrides the base method to calculate salary including the bonus.
        """
        # Get the base salary from the parent method and add the bonus
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example Usage:

# Create an Employee instance
employee1 = Employee("Vijay", hours_worked=40, hourly_rate=20)
print(f"{employee1.name}'s salary: ${employee1.calculate_salary():.2f}")

# Create a Manager instance
manager1 = Manager("Vikki", hours_worked=40, hourly_rate=25, bonus=1000)
print(f"{manager1.name}'s salary: ${manager1.calculate_salary():.2f}")

```

#15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product

```python
class Product:
    """
    Represents a product with a name, price, and quantity.
    """
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        """
        Calculates and returns the total price of the product instance
        (price per unit multiplied by quantity).
        """
        return self.price * self.quantity

# --- Example Usage ---

# Create an instance of the Product class: a dozen eggs costing 2.50 each
egg_carton = Product(name="Eggs (Dozen)", price=2.50, quantity=12)

# Call the total_price method for this specific instance
total_cost = egg_carton.total_price()

# Print the results
print(f"Product: {egg_carton.name}")
print(f"Unit Price: ${egg_carton.price:.2f}")
print(f"Quantity: {egg_carton.quantity}")
print(f"Total Cost: ${total_cost:.2f}")
```

# 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

```python
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class 1: Cow
class Cow(Animal):
    def sound(self):
        return "Moo"

# Derived class 2: Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa"

# Demonstrating functionality
if __name__ == "__main__":
    cow = Cow()
    sheep = Sheep()

    print(f"Cow says: {cow.sound()}")
    print(f"Sheep says: {sheep.sound()}")
```

```python
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author} ({self.year_published})"

# Example usage:
my_book = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)
print(my_book.get_book_info())
```

# 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

```python
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        # Use super() to initialize attributes from the House class
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Example usage:
my_mansion = Mansion("123 Luxury Lane", 5000000, 12)
print(f"Address: {my_mansion.address}, Rooms: {my_mansion.number_of_rooms}")
```
