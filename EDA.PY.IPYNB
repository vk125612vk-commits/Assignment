Question 1: What is Simple Linear Regression?


Answer:
         Simple Linear Regression is a basic statistical method to model the straight-line relationship between one independent variable (X) and one dependent variable (Y), aiming to predict Y from X by finding the "line of best fit" (like \(Y=a+bX\)) that minimizes errors in the data, helping understand how much Y changes with X


Question 2: What are the key assumptions of Simple Linear Regression?

Answer: The key assumptions for simple linear regression are Linearity (straight-line relationship), Independence (errors aren't related), Homoscedasticity (constant error variance), and Normality (errors are normally distributed), along with the expectation of Zero Mean Error, ensuring the model accurately describes the data for valid inference. These conditions ensure the estimated line truly represents the average Y for given X values. 

Key Assumptions of Simple Linear Regression
1. Linearity
The relationship between the independent and dependent variables is linear.

The dependent variable should change proportionally with the independent variables, forming a straight-line trend.
Curved or irregular patterns can cause underfitting and inaccurate predictions.
When linearity fails, data transformations or non-linear models may be required.

2. Homoscedasticity of Residuals
The variance of residuals remains constant across all levels of the independent variables.

Residuals should appear evenly scattered, indicating uniform error spread.
Patterns of increasing or decreasing variance lead to unreliable coefficient estimates.
Severe heteroscedasticity may require transformations or weighted regression methods.

3. Multivariate Normality - Normal Distribution
The residuals follow a normal distribution when multiple predictors are involved.

Normality supports valid confidence intervals, hypothesis tests and p-values.
Skewed or peaked distributions weaken inference quality.
Violations may be corrected through transformation or larger sample sizes.

4. Independence of Errors
Residuals must not correlate with each other across observations.

Correlated errors suggest the model missed temporal or patterned structure.
Autocorrelation can inflate significance and mislead conclusions.
Time-series data often require specialized methods to resolve this.

5. Lack of Multicollinearity
The independent variables are not highly correlated with each other.

Strong collinearity inflates coefficient variance and reduces interpretability.
It becomes difficult to assess the true contribution of each predictor.
Feature selection or regularization helps reduce the effect.

6. Absence of Endogeneity
Independent variables in the regression model should not be correlated with the error term.

Endogeneity causes biased and inconsistent parameter estimates.
Inference based on such coefficients becomes unreliable.
Instrumental variables or additional predictors can address this issue.


Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer: Definition: From Greek words for "different" (hetero) and "dispersion" (skedasis), it's when the variance of the error term (residuals) systematically changes as the independent variable(s) change, unlike homoscedasticity (constant variance).

Visualized: In a residual plot, this looks like a funnel shape (errors widen or narrow as X increases) or a changing pattern, rather than a consistent band of points.

Example: Income vs. food spending; a poor person's spending is consistent, but a wealthy person's varies widely (high income = high variance)

Why It's Important to Address


Unreliable Standard Errors: OLS assumes constant variance; heteroscedasticity makes standard errors incorrect, often underestimating them.


Invalid Inference: Incorrect standard errors lead to wrong t-statistics and p-values, meaning you might think a variable is significant when it isn't, or vice-versa.

Inefficient Estimates: Coefficients remain unbiased but are no longer the "best" (most efficient) estimates, losing precision.

Poor Predictions: The model's predictive power can be unreliable, especially for extreme values where the error spread is largest. 


Question 4: What is Multiple Linear Regression?

Answer: Multiple Linear Regression (MLR) is a statistical method to predict a single dependent variable (outcome) using the values of two or more independent variables (predictors), finding the best-fit linear equation that describes their combined relationship to forecast results, such as predicting a car's price (dependent) based on its age, mileage, and color (independent). It's an extension of simple linear regression, allowing researchers to see how multiple factors influence an outcome simultaneously while holding others constant. [1, 2, 3]  
How it Works 

• Equation: The model finds coefficients (weights) for each predictor, creating a formula like: $Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p + \epsilon$, where Y is the outcome, $X_i$ are predictors, $\beta_0$ is the intercept, $\beta_i$ are coefficients, and $\epsilon$ is the error. 
• Line of Best Fit: It fits a "line" (or hyperplane in higher dimensions) to the data, minimizing the sum of squared errors (differences between actual and predicted values). 
• Variable Importance: It helps determine which independent variables significantly contribute to the prediction and their relative impact. [1, 2, 4, 5, 6, 7]  

Key Assumptions 

• Linearity: A linear relationship exists between variables. 
• Independence: Observations are independent. 
• Homoscedasticity: Residuals (errors) have constant variance. 
• No Multicollinearity: Independent variables are not too highly correlated with each other. [8]  

Example 
Predicting a student's test score (Dependent) using hours studied, prior GPA, and anxiety level (Independent). MLR reveals how much each factor, when others are controlled for, impacts the score, and how well the model predicts scores overall (R-squared). [1, 5, 9, 10]  



Question 5: What is polynomial regression, and how does it differ from linear regression?

Answers:

 Polynomial regression is a regression technique that uses polynomial terms to model the non-linear relationship between the dependent variable (Y) and the independent variable (X).


```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4])
Y = np.array([2, 5, 10, 17])

# Polynomial regression (degree = 2)
coefficients = np.polyfit(X, Y, 2)
poly_model = np.poly1d(coefficients)

# Predicted values
X_new = np.linspace(1, 4, 100)
Y_new = poly_model(X_new)

# Plot
plt.scatter(X, Y)
plt.plot(X_new, Y_new)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression")
plt.show()
```

Question 6 Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.
(Include your Python code and output in the code box below.)

Answers:

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit Simple Linear Regression model
# y = a + bX
b, a = np.polyfit(X, Y, 1)

# Predicted values
Y_pred = a + b * X

# Plot the data points and regression line
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()

print("Intercept (a):", a)
print("Slope (b):", b)
```

Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results. (Include your Python code and output in the code box below.)

Answers:

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Given data
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

# Create DataFrame
df = pd.DataFrame({
    'Area': Area,
    'Rooms': Rooms,
    'Price': Price
})

# Independent and dependent variables
X = df[['Area', 'Rooms']]
y = df['Price']

# Add constant
X = sm.add_constant(X)

# Fit Multiple Linear Regression model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())

# Calculate VIF
vif = pd.DataFrame()
vif['Variable'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i)
              for i in range(X.shape[1])]

print(vif)
```

Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.
(Include your Python code and output in the code box below.)

Answers:

```python
# Polynomial Regression (2nd Degree)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Convert X to polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Model output
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Create smooth curve
X_curve = np.linspace(1, 5, 100).reshape(-1, 1)
X_curve_poly = poly.transform(X_curve)
Y_curve = model.predict(X_curve_poly)

# Plot
plt.figure()
plt.scatter(X, Y)
plt.plot(X_curve, Y_curve)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.
(Include your Python code and output in the code box below.)

Answers

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])

# Fit simple linear regression using numpy
coefficients = np.polyfit(X, Y, 1)
model = np.poly1d(coefficients)

# Predicted values
Y_pred = model(X)

# Residuals
residuals = Y - Y_pred

# Print outputs
print("Regression coefficients:", coefficients)
print("Predicted Y:", Y_pred)
print("Residuals:", residuals)

# Residuals plot
plt.figure()
plt.scatter(X, residuals)
plt.axhline(0)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

Answers: Picture the model as a house itself 🏠. If the foundation is uneven or the walls lean into each other, no amount of paint will save it. Heteroscedasticity and multicollinearity are exactly those structural problems. Here is how I would fix them, step by step, as a data scientist in a real estate company.

---

## 1️⃣ First: Diagnose clearly (don’t guess)

### Heteroscedasticity

* Plot **residuals vs predicted prices**
* Look for a “fan” or “cone” shape where residual spread increases with price
* Run formal tests like **Breusch–Pagan** or **White test**

### Multicollinearity

* Check **correlation matrix**
* Calculate **VIF (Variance Inflation Factor)**

  * VIF > 5 (or 10) is a red flag
    Example in real estate: *area* and *number of rooms* moving almost together

---

## 2️⃣ Fixing Heteroscedasticity (uneven error spread)

### a) Transform the target variable

* Apply **log(price)** or **√price**
* Very common in housing data because price variance grows with size and location quality

👉 This often stabilizes variance naturally

### b) Use robust standard errors

* Keep coefficients but correct inference using **heteroscedasticity-robust (HC) errors**
* Ensures reliable p-values and confidence intervals

### c) Weighted Least Squares (WLS)

* Give **less weight to high-variance observations**
* Useful when expensive luxury homes dominate the error spread

---

## 3️⃣ Fixing Multicollinearity (predictors stepping on each other)

### a) Feature selection

* Remove redundant variables
  Example:

  * Keep **area**, drop **rooms**, or vice versa
* Choose based on business meaning and predictive power

### b) Feature engineering

* Combine correlated features:

  * `price_per_sqft = price / area`
  * `rooms_per_1000sqft`
* This reduces overlap and improves interpretability

### c) Regularization techniques

* **Ridge Regression** → shrinks coefficients, great for multicollinearity
* **Lasso Regression** → can eliminate weak predictors entirely
* **Elastic Net** → best of both worlds

---

## 4️⃣ Location-specific handling (real estate reality)

* Encode location smartly:

  * Target encoding or grouped location dummies
* Consider **separate models** or interaction terms for premium vs non-premium areas
* This reduces hidden multicollinearity caused by location effects

---

## 5️⃣ Re-evaluate and validate

After fixes:

* Re-check **residual plots**
* Recalculate **VIF**
* Validate using **cross-validation**
* Compare performance using **RMSE / MAE**

A model that performs well on unseen data is the real “price tag” of success 💰📈

---

## 6️⃣ Final takeaway (exam-ready)

> Heteroscedasticity is handled using transformations, robust errors, or weighted regression, while multicollinearity is addressed through feature selection, regularization, and feature engineering. These steps improve model stability, interpretability, and predictive reliability.
