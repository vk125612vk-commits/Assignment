# Question 1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Answer:

In hypothesis testing:

* Type I Error (False Positive)
Rejecting the null hypothesis when it is actually true.
Probability of Type I error = Î± (significance level).

* Type II Error (False Negative)
Failing to reject the null hypothesis when it is actually false.
Probability of Type II error = Î².

Impact on decision-making:

* A Type I error may lead to unnecessary actions or incorrect conclusions.

* A Type II error may result in missed opportunities or failure to detect real effects.

* There is a trade-off between Î± and Î²; reducing one often increases the other.

# Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Answer:

Purpose:
ANOVA (Analysis of Variance) tests whether there are significant differences between the means of three or more groups.

Assumptions:

1. Independence of observations

2. Normal distribution of groups

3. Homogeneity of variances

Extension:
Instead of performing multiple t-tests, ANOVA compares all group means simultaneously using variance ratios (F-statistic).

# Question 6: Write a Python program to perform a one-sample Z-test and interpret the result.



```python
import numpy as np
from scipy import stats

# Sample data
data = [52, 55, 50, 53, 54, 56, 52, 51]
population_mean = 50
population_std = 5

sample_mean = np.mean(data)
n = len(data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 1 - stats.norm.cdf(z_score)

print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

# Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Non-normal distribution (Exponential)
population = np.random.exponential(scale=2, size=10000)

sample_means = []

for i in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot distribution of sample means
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

# Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2., len(data)-1)

    lower = mean - margin
    upper = mean + margin

    print("Confidence Interval:", (lower, upper))

    plt.errorbar(0, mean, yerr=margin, fmt='o')
    plt.title("Confidence Interval")
    plt.show()

# Example
data = np.random.normal(50, 5, 30)
confidence_interval(data)
```

# Question 10: Perform a Chi-square goodness-of-fit test using Python and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: Observed distribution differs from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```
