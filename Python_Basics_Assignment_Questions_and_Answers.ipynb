#2. What is an interpreter in Python?


Answer:

An interpreter is a program that reads Python code line by line and executes it immediately. Unlike compiled languages, Python doesn't need to be compiled before running - the interpreter handles execution directly.



#3. What are pre-defined keywords in Python?


Answer:

Keywords are reserved words in Python that have special meanings and cannot be used as identifiers (variable names, function names, etc.). Examples: if, else, while, for, def, class, import, True, False, None.



#4. Can keywords be used as variable names?


Answer:

No, keywords cannot be used as variable names. Attempting to do so will cause a SyntaxError.



#5. What is mutability in Python?


Answer:

Mutability refers to whether an object can be modified after creation. Mutable objects can be changed, immutable objects cannot.



#6. Why are lists mutable, but tuples are immutable?


Answer:

*   Lists are mutable because they're designed to store collections of items that may need to be modified (added, removed, changed)


*   Tuples are immutable because they're meant to store fixed collections of items, providing data integrity and allowing them to be used as dictionary keys

#9. What is type casting in Python?


Answer:

Type casting is the process of converting one data type to another. For example: converting a string to an integer.



#10. What is the difference between implicit and explicit type casting?


Answer:


> * Implicit: Python automatically converts types (e.g., 3 + 4.5 converts 3 to 3.0)

> * Explicit: Programmer manually converts types using functions like int(), str(), float()

#13. What is the difference between for and while loops?


Answer:

#14. Describe a scenario where a while loop is more suitable than a for loop


Answer:

A while loop is better when you don't know in advance how many iterations are needed. Example: Reading user input until a specific value is entered:

```python
user_input = "Vijay"
while user_input != "quit":
    user_input = input("Enter something (type 'quit' to exit): ")
    print(f"You entered: {user_input}")
```
