#Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:

Anomaly Detection is the process of identifying patterns or data points that deviate significantly from the expected behavior or normal distribution of a dataset. These deviations, called anomalies or outliers, often indicate rare events, errors, fraud, or system faults.

> Types of Anomalies:


 1. Point Anomaly – An individual data point that is anomalous relative to the rest of the data.
Example: In credit card transaction data, a single transaction of $10,000 when the user typically spends $50–$100 per transaction.

1. Contextual Anomaly – A data point that is anomalous within a specific context (e.g., time, location) but not otherwise.
Example: A temperature of 30°C (86°F) is normal in summer but anomalous in winter for a particular city. The anomaly depends on the seasonal context.

3. Collective Anomaly – A collection of related data points that together form an anomalous pattern, even if individually each point may be normal.
Example: A cyber‑attack sequence – several login attempts from different IP addresses in a short time, each normal individually but collectively indicating a brute‑force attack.



# Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases

Answer:

#🔍 1. Isolation Forest (iForest)

Approach (Tree-based, isolation logic):
Instead of profiling “normal” points, it isolates anomalies.

* Randomly splits data using decision trees
* Outliers get isolated faster (shorter path length)

#ntuition:

* Rare points are easier to separate than common ones.

#Strengths:

* Works well on high-dimensional data
* Fast and scalable (great for big datasets 🚀)
* No need for distance or density calculation

#Weakness:

* Doesn’t capture local density variations well

#Best Use Cases:

* Fraud detection (banking, fintech)
* Network intrusion detection
* Large datasets with many features

🌌 2. DBSCAN (Density-Based Spatial Clustering)

Approach (Density-based clustering):

Groups points into clusters based on density.

* Points in low-density regions = anomalies (noise)

#Key Idea:

Normal data = dense clusters

Outliers = points floating alone in sparse space

#Strengths:

* Can detect arbitrary-shaped clusters
* Automatically identifies outliers (no separate step)
* No need to specify number of clusters

#Weakness:

* Sensitive to parameters (eps, min_samples)

* Struggles with varying densities

#Best Use Cases:

* Geospatial data (e.g., GPS, location clustering 🗺️)

* Image segmentation

* Noise detection in clustering problems

#🧭 3. Local Outlier Factor (LOF)

Approach (Density-based, local comparison):

Compares density of a point with its neighbors.

* If a point has much lower density than neighbors → outlier

#Key Idea:
* “Is this point weird relative to its neighborhood?”

#Strengths:

* Captures local anomalies
* Works well when data has varying density

#Weakness:

* Computationally expensive (distance calculations)
* Not great for very large datasets

#Best Use Cases:

* Credit card fraud (local unusual behavior)
* Sensor data anomalies
* Detecting subtle deviations in small datasets

#Question 3: What are the key components of a Time Series? Explain each with one example.


Answer:

A time series typically consists of four key components:

>1. Trend – Long‑term increase or decrease in the data over time.
Example: Global average temperature rising over decades due to climate change.

>2. Seasonality – Regular, predictable patterns that repeat at fixed intervals (daily, weekly, monthly, yearly).
Example: Ice cream sales increase every summer and decrease every winter.

>3. Cyclical – Fluctuations that occur at irregular intervals, often influenced by economic or business cycles (longer than seasonal).
Example: Stock market bull and bear cycles lasting several years.

>4. Irregular (or Residual) – Random, unpredictable variations not explained by trend, seasonality, or cycles.
Example: Sudden drop in retail sales due to an unexpected snowstorm.



# Question 4: Define Stationary in time series. How can you test and transform a non‑stationary series into a stationary one?


Answer:

Stationarity means that the statistical properties of a time series (mean, variance, autocorrelation) do not change over time. A stationary series has no trend, constant variance, and a constant autocorrelation structure.

Tests for Stationarity:

>*  Augmented Dickey‑Fuller (ADF) test – Null hypothesis: series is non‑stationary. Rejecting H0 indicates stationarity.

>* Kwiatkowski‑Phillips‑Schmidt‑Shin (KPSS) test – Null hypothesis: series is stationary.

Transformations to achieve stationarity:

1. Differencing – Compute the difference between consecutive observations:yt′​=yt​−yt−1​.
Apply second‑order differencing if needed.

2. Log transformation – Stabilizes variance, especially when variance increases with level.

3. Detrending – Subtract a fitted trend line (linear, polynomial, etc.).

4. Seasonal differencing – Remove seasonal patterns:
y  t ′ ​  =y  t ​  −y  t−s ​

8 is the seasonal period.

5. Power transforms – Box‑Cox transform to stabilize variance.

#Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.


Answer:

| Model | Structure | Application |
|-------|-----------|-------------|
| **AR (AutoRegressive)** | \( y t ​  =c+ϕ  1 ​  y  t−1 ​  +⋯+ϕ  p ​  y  t−p ​  +ε t ​) | Stationary series with dependence on past values (e.g., stock returns). |
| **MA (Moving Average)** | \( y  t ​  =c+ε  t ​  +θ  1 ​  ε  t−1 ​  +⋯+θ  q ​  ε  t−q ​) | Series influenced by past forecast errors (e.g., white noise with memory). |
| **ARIMA (p,d,q)** | AR + MA + differencing (d) to handle non‑stationarity. | General univariate time series without seasonality (e.g., interest rates). |
| **SARIMA (p,d,q)(P,D,Q,s)** | ARIMA + seasonal components (P,D,Q) with seasonal period s. | Univariate series with strong seasonal patterns (e.g., monthly sales, airline passengers). |
| **SARIMAX** | SARIMA + exogenous variables (X) – external predictors. | Seasonal series influenced by external factors (e.g., electricity demand + temperature, holidays). |



# Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

```python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset from the CSV file
# Assuming the file 'AirPassengers.csv' is in the current working directory
df = pd.read_csv('AirPassengers.csv')

# Display first few rows to confirm structure
print("First 5 rows of the dataset:")
print(df.head())

# Convert 'Month' column to datetime and set as index
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# Rename the passengers column for clarity
df.columns = ['Passengers']

# Plot the original time series
plt.figure(figsize=(12, 5))
plt.plot(df.index, df['Passengers'], color='blue', linewidth=1.5)
plt.title('Monthly Air Passengers (1949–1960)')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()

# Decompose the series (multiplicative model because variance increases with trend)
# Period = 12 for monthly seasonality
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)

# Extract components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot all components
fig, axes = plt.subplots(4, 1, figsize=(12, 10))

df['Passengers'].plot(ax=axes[0], title='Original Series', color='blue', linewidth=1.5)
axes[0].set_ylabel('Passengers')

trend.plot(ax=axes[1], title='Trend Component', color='green', linewidth=1.5)
axes[1].set_ylabel('Passengers')

seasonal.plot(ax=axes[2], title='Seasonal Component', color='orange', linewidth=1.5)
axes[2].set_ylabel('Seasonal Factor')

residual.plot(ax=axes[3], title='Residual Component', color='red', linewidth=1.5)
axes[3].set_ylabel('Residuals')

plt.tight_layout()
plt.show()

# Print summary statistics of the residual to check randomness
print("\nResidual summary statistics (should be close to 1 for multiplicative model):")
print(residual.describe())
```

#Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load the dataset
df = pd.read_csv('NYC_taxi_fare_data.csv')

# Display basic info
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())

# Select numerical features for anomaly detection
# Using trip_distance (miles) and fare_amount (USD)
data = df[['trip_distance', 'fare_amount']].copy()

# Remove rows with missing values (if any)
data.dropna(inplace=True)

# Basic data cleaning: filter out obviously erroneous or extreme values
# Trip distance should be >= 0 and reasonable (e.g., <= 20 miles for typical NYC taxi)
# Fare amount should be > 0 and not excessively high (e.g., <= 100)
data = data[(data['trip_distance'] >= 0) & (data['trip_distance'] <= 20)]
data = data[(data['fare_amount'] > 0) & (data['fare_amount'] <= 100)]

print(f"\nData points after cleaning: {len(data)}")

# Initialize Isolation Forest
# contamination=0.1 assumes about 10% anomalies (can be adjusted)
iso_forest = IsolationForest(contamination=0.1, random_state=42)

# Fit the model and predict anomalies
data['anomaly'] = iso_forest.fit_predict(data[['trip_distance', 'fare_amount']])

# Isolation Forest returns -1 for anomalies, 1 for normal points
anomalies = data[data['anomaly'] == -1]
normals = data[data['anomaly'] == 1]

print(f"\nNormal points: {len(normals)}")
print(f"Anomalies detected: {len(anomalies)} ({len(anomalies)/len(data)*100:.2f}%)")

# Visualize the anomalies on a 2D scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(normals['trip_distance'], normals['fare_amount'],
            c='blue', label='Normal', alpha=0.5, s=20)
plt.scatter(anomalies['trip_distance'], anomalies['fare_amount'],
            c='red', label='Anomaly', alpha=0.7, s=40, marker='x')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.title('Isolation Forest Anomaly Detection on NYC Taxi Fare Data')
plt.legend()
plt.grid(True)
plt.show()
```

# Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load AirPassengers dataset
data = sm.datasets.get_rdataset("AirPassengers").data
data.columns = ['time', 'passengers']
data['time'] = pd.date_range(start='1949-01-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)

# Train/validation split (use data up to end of 1959 for training)
train = data['passengers'][:12*11]  # 1949-01 to 1959-12 (132 months)
test = data['passengers'][12*11:]   # 1960-01 to 1960-12 (12 months)

# Fit SARIMA model (order chosen based on ACF/PACF and known best practices)
# Seasonal period = 12 (monthly)
model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)
print(results.summary())

# Forecast next 12 months
forecast = results.forecast(steps=12)
forecast_index = pd.date_range(start='1960-01-01', periods=12, freq='M')
forecast_series = pd.Series(forecast, index=forecast_index)

# Visualize
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data (1949-1959)', color='blue')
plt.plot(test.index, test, label='Actual Test (1960)', color='green', linestyle='--')
plt.plot(forecast_series.index, forecast_series, label='SARIMA Forecast (12 months)', color='red', marker='o')
plt.fill_between(forecast_series.index,
                 forecast_series - 1.96*np.std(forecast),
                 forecast_series + 1.96*np.std(forecast),
                 color='red', alpha=0.2, label='95% Confidence Interval')
plt.title('SARIMA Model Forecast for Monthly Air Passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate forecast error
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}")
```

# Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Load a numerical dataset – using NYC Taxi Fare again (or a synthetic dataset)

df = pd.read_csv("/content/NYC_taxi_fare_data.csv")

# Use two features: fare_amount and trip_distance
data = df[['fare_amount', 'trip_distance']].dropna()
data = data[(data['fare_amount'] > 0) & (data['fare_amount'] < 100)]
data = data[(data['trip_distance'] > 0) & (data['trip_distance'] < 30)]

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
data['anomaly'] = lof.fit_predict(data[['fare_amount', 'trip_distance']])
# LOF returns -1 for outliers, 1 for inliers
anomalies = data[data['anomaly'] == -1]
normals = data[data['anomaly'] == 1]

# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(normals['trip_distance'], normals['fare_amount'],
            c='blue', label='Normal', alpha=0.5, s=10)
plt.scatter(anomalies['trip_distance'], anomalies['fare_amount'],
            c='red', label='Anomaly (LOF)', alpha=0.7, s=30)
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.title('Local Outlier Factor (LOF) Anomaly Detection on NYC Taxi Fare Data')
plt.legend()
plt.grid(True)
plt.show()

print(f"Total points: {len(data)}")
print(f"Anomalies detected by LOF: {len(anomalies)} ({len(anomalies)/len(data)*100:.2f}%)")
```

# Question 10: You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real‑time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage. Explain your real‑time data science workflow.




**Answer:**

### Real‑Time Data Science Workflow for Power Grid Monitoring

#### 1. Anomaly Detection in Streaming Data

**Choice: Isolation Forest (with sliding window)**  
- **Why Isolation Forest?**  
  - Efficient for high‑dimensional streaming data (timestamp, region, weather, usage).  
  - Handles large throughput (15‑minute intervals from thousands of sensors).  
  - Can be updated incrementally using a sliding window (e.g., last 7 days of data) to adapt to daily/seasonal patterns.  
- **Alternative (not chosen):**  
  - DBSCAN is less suitable because cluster densities vary with time of day.  
  - LOF is computationally heavier for real‑time streaming.  

**Implementation:**  
- Buffer the last `N` observations (e.g., 672 points = 7 days × 96 quarters).  
- Train Isolation Forest on this window every hour (or use online variant).  
- For each new 15‑minute reading, score the point; if anomaly score < threshold, raise alert.  
- Use contextual anomaly detection by adding engineered features: hour_of_day, day_of_week, holiday_flag.

#### 2. Short‑Term Forecasting

**Choice: SARIMAX (with weather as exogenous variable)**  
- **Why SARIMAX?**  
  - Energy demand has strong daily (24‑hour) and weekly seasonality → SARIMA handles this.  
  - Weather conditions (temperature, humidity) significantly affect demand → exogenous variables (X) improve accuracy.  
- **Forecast horizon:** Next 24–48 hours (96–192 steps ahead).  

**Implementation:**  
- Re‑train model daily using last 30 days of data.  
- Include weather forecasts as exogenous inputs for future steps.  
- Output probabilistic forecasts (prediction intervals) to capture uncertainty.

#### 3. Validation and Monitoring Over Time

- **Backtesting:** Use rolling window validation – simulate real‑time by training on past data and testing on next week.  
- **Metrics:**  
  - Forecasting: MAE, RMSE, MAPE, and pinball loss for quantile forecasts.  
  - Anomaly detection: Precision/Recall at different thresholds (use labeled historical events).  
- **Monitoring dashboards:**  
  - Track model drift (e.g., change in residual distribution).  
  - Alert when false positive rate exceeds 5% or detection latency > 2 minutes.  
- **Retraining strategy:**  
  - Automated daily retraining for forecasting model.  
  - Hourly update of Isolation Forest window (without full retraining).

#### 4. Business Impact and Operational Decisions

- **Grid stability:** Early detection of spikes (e.g., sudden industrial load) allows operators to reroute power or activate reserve generation, preventing blackouts.  
- **Demand response:** Accurate forecasts enable utilities to negotiate real‑time pricing, incentivize consumers to shift usage, and reduce peak loads.  
- **Maintenance scheduling:** Detecting abnormal drops (e.g., a substation failure) triggers immediate inspection and reduces downtime.  
- **Regulatory compliance:** Automated anomaly logs help meet reliability standards (e.g., NERC CIP).  
- **Cost savings:** Reduced need for expensive peaker plants, lower energy procurement costs, and avoided infrastructure damage.

**Conclusion:** The combination of **Isolation Forest for real‑time anomaly detection** and **SARIMAX for short‑term forecasting** provides a robust, scalable solution for smart grid monitoring. Continuous validation ensures the system adapts to changing consumption patterns and weather conditions.
