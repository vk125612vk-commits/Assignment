#4 Write a program that checks if a given word is a Python keyword.

```python
import keyword

def check_keyword(word):
    if keyword.iskeyword(word):
        print(f"'{word}' is a Python keyword")
    else:
        print(f"'{word}' is not a Python keyword")

# Example usage
check_keyword("if")
check_keyword("hello")
```

#6 Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def demonstrate_mutability(data, number):

    print(f"\nInside function - before modification:")
    print(f"Data: {data}, Number: {number}")

    # Try to modify
    if isinstance(data, list):
        data.append(99)  # Modifies the original list
    number = 999         # Creates a new local variable

    print(f"Inside function - after modification:")
    print(f"Data: {data}, Number: {number}")

# Test with mutable (list) and immutable (int)
my_list = [1, 2, 3]
my_num = 5

print(f"Before function call: {my_list}, {my_num}")
demonstrate_mutability(my_list, my_num)
print(f"After function call: {my_list}, {my_num}")
```

#7. Write a program that performs basic arithmetic operations on two user-input numbers.

```python
def arithmetic_operations():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    print(f"\n{num1} + {num2} = {num1 + num2}")
    print(f"{num1} - {num2} = {num1 - num2}")
    print(f"{num1} * {num2} = {num1 * num2}")

    if num2 != 0:
        print(f"{num1} / {num2} = {num1 / num2}")
        print(f"{num1} % {num2} = {num1 % num2}")
    else:
        print("Cannot divide by zero")

# Run the function
arithmetic_operations()
```

#9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
def type_conversion():
    user_input = input("Enter a value: ")

    # Convert to different types
    try:
        int_value = int(user_input)
        print(f"As integer: {int_value} (type: {type(int_value)})")
    except ValueError:
        print("Cannot convert to integer")

    try:
        float_value = float(user_input)
        print(f"As float: {float_value} (type: {type(float_value)})")
    except ValueError:
        print("Cannot convert to float")

    # Boolean conversion
    bool_value = bool(user_input)
    print(f"As boolean: {bool_value} (type: {type(bool_value)})")

type_conversion()
```

# 10. Write code to demonstrate type casting with list elements.

```python
import numpy as np
from sklearn.ensemble import IsolationForest

# Generate data
# normal cluster around 0
normal = np.random.randn(60, 1)

# anomaly cluster around 8
anomalies = np.random.randn(40, 1) + 8

X = np.vstack([normal, anomalies])

# Case 1: treat ~10% as anomalies
model_low = IsolationForest(contamination=0.1, random_state=42)
model_low.fit(X)
pred_low = model_low.predict(X)
print("Case 10% contamination -> flagged anomalies:", np.sum(pred_low == -1))

# Case 2: treat ~40% as anomalies
model_high = IsolationForest(contamination=0.4, random_state=42)
model_high.fit(X)
pred_high = model_high.predict(X)
print("Case 40% contamination -> flagged anomalies:", np.sum(pred_high == -1))
```

# 13. Write a Python program to find the sum of all even numbers between 1 and 50.


```python
# Program to find the sum of even numbers between 1 and 50

total = 0

for number in range(1, 51):
    if number % 2 == 0:   # Check if the number is even
        total += number

print("Sum of even numbers between 1 and 50 is:", total)
```

# 14. Write a program to reverse a string using a while loop.

```python
# Program to reverse a string using a while loop

text = input("Enter a string: ")

reversed_text = "Vijay"
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)
```

#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Program to calculate factorial using a while loop

number = int(input("Enter a number: "))

factorial = 1
i = 1

# Check for negative numbers
if number < 0:
    print("Factorial does not exist for negative numbers.")
else:
    while i <= number:
        factorial *= i
        i += 1

    print("Factorial of", number, "is:", factorial)
```
