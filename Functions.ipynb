# 1. What is the difference between a function and a method in Python?

Answer:

In Python, the primary difference is that a function is a standalone block of code, while a method is a function that is associated with an object or a class. All methods are functions, but not all functions are methods.


* Function: A standalone block of reusable code that performs a specific task. It is defined using def and called independently (e.g., print(), len()).

* Method: A function that is associated with an object and is called on that object. It is defined inside a class and operates on the data of that instance (e.g., my_list.append(5), my_string.upper()).

```python
# Function (standalone)
def greet(name):
    return f"Hello, {name}!"

# Method (inside a class)
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        # Method accesses the object's attribute (self.name)
        return f"{self.name} says woof!"

# Function call
print(greet("Vijay"))

# Method call (on an object instance)
my_dog = Dog("Rex")
print(my_dog.bark())
```

# 5 What are iterators in Python and how do they differ from iterables?

Answer:

üîπ What is an Iterable?

An iterable is any object that can be looped over.

Examples:

* list

* tuple

* string

* dictionary

* set

```python
my_list = [1, 2, 3]
for i in my_list:
    print(i)
```

üîπ What is an Iterator?

An iterator is an object that:

* Keeps track of current position

* Produces the next value using __next__()

* Raises StopIteration when finished

You create an iterator using iter():

```python
my_list = [1, 2, 3]
my_iterator = iter(my_list)

print(next(my_iterator))  # 1
print(next(my_iterator))  # 2
print(next(my_iterator))  # 3
```

| Iterable             | Iterator                          |
| -------------------- | --------------------------------- |
| Contains data        | Produces data one by one          |
| Can be looped over   | Keeps state of iteration          |
| Has `__iter__()`     | Has `__iter__()` and `__next__()` |
| Example: list, tuple | Example: object from `iter()`     |

#üîπ Example to See the Difference


```python
my_list = [10, 20, 30]

print(hasattr(my_list, "__next__"))  # False

it = iter(my_list)
print(hasattr(it, "__next__"))  # True
```

#6. Explain the concept of generators in Python and how they are defined.


Answer:

‚Äù

üîπ What is a Generator?

A generator is a special type of iterator that:

Produces values lazily (on demand)

Saves memory

Remembers its state between calls

Uses the yield keyword instead of return.




üîπ How Generators Are Defined

There are two main ways to define generators.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# üîπ Using it:

counter = count_up_to(3)

print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
```

2Ô∏è‚É£ Generator Expression (Compact Version)

Similar to list comprehension, but with parentheses.

```python
squares = (x * x for x in range(5))

#üîπ Using it:

print(next(squares))  # 0
print(next(squares))  # 1
```

This does NOT create a full list in memory.

Compare:

```python
[x * x for x in range(5)]   # List (stored fully)
(x * x for x in range(5))   # Generator (lazy)
```

# 7. What are the advantages of using generators over regular functions?

Answer:


üîπ Advantages of Generators Over Regular Functions

1Ô∏è‚É£ Memory Efficiency

Regular function with list:

```python
def numbers():
    return [i for i in range(1000000)]
```

This creates 1 million numbers in memory at once.

Generator version:

```python
def numbers():
    for i in range(1000000):
        yield i
```

This produces one number at a time.

üìå Perfect for large datasets, big files, or streaming data.

2Ô∏è‚É£ Lazy Evaluation (On-Demand Execution)

Generators calculate values only when needed.

```python
gen = (x * 2 for x in range(5))

#Nothing runs until you call:

next(gen)

#This makes programs faster and more efficient.
```

3Ô∏è‚É£ Maintains State Automatically

A generator remembers where it stopped.

```python
def counter():
    yield 1
    yield 2
    yield 3

  # Each next() continues from the last pause point.

  # A normal function starts fresh every time.
```

4Ô∏è‚É£ Cleaner Code for Large Pipelines

Generators work beautifully in data pipelines:

```python
def read_file(file):
    for line in file:
        yield line.strip()
```

Used in:

* File processing

* Machine learning preprocessing

* Log analysis

* API data streaming

Less memory. Cleaner flow. Smoother performance


5Ô∏è‚É£ Infinite Sequences Possible

Regular functions must finish.
Generators can run forever.

```python
def infinite_numbers():
    i = 1
    while True:
        yield i
        i += 1
```

Same playground. Different superpowers.

It applies a function to each element of an iterable.

‚úÖ Purpose:

Modify or transform data

```python
  numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, numbers))
print(result)
```


üîπ 2Ô∏è‚É£ filter() ‚Üí Select Some Items

It keeps only elements where the condition is True.

‚úÖ Purpose:

Filter data based on a condition.

Aggregate values into one result.

# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for i in range(1,201):
    if is_prime(i):
        print(i)
```

# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(10):
    print(num)
```

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

for val in powers_of_two(5):
    print(val)
```

# 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
```
